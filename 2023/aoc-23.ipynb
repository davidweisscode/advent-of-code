# 01 Trebuchet

```python
import re

def get_numerical_str(line):
    line = re.sub("one", "1", line)
    line = re.sub("two", "2", line)
    line = re.sub("three", "3", line)
    line = re.sub("four", "4", line)
    line = re.sub("five", "5", line)
    line = re.sub("six", "6", line)
    line = re.sub("seven", "7", line)
    line = re.sub("eight", "8", line)
    line = re.sub("nine", "9", line)
    return line
```

```python
with open("input.txt", "r+") as file:
    lines = file.read().splitlines()

sum = 0

for line in lines:
    original_line = line
    if len(original_line) > 4:
        substr_numerical = []
        for i in range(len(original_line) - 4):
            substr = original_line[i:i+5]
            substr_numerical.append(get_numerical_str(substr))
        line = "".join(substr_numerical)
    line = re.sub("[^0-9]", "", line)
    first = line[0]
    last = line[-1]
    digits = first + last
    sum += int(digits)
    print(original_line, line, first, last, digits, "\n+++")

print("output", sum)
```
